```python
import pandas as pd
import operator
import functools
```

```python
search = pd.read_excel("Edited.xlsx", sheet_name=1)
search.head()
```

```python
firms = pd.read_csv("firm - firm.csv", sep=";", names=["company", "date", "price"])
firms.head()
```

```python
request = pd.read_csv("request - request.csv", header=None, names=["event", "company", "market", "date", "type", "l", "r", "w", "c"])
request.to_csv("request - request;.csv", sep=";", header=False, index=False)
request.head()
request.assign(l=-1, r=1, w=-2, c=4).to_csv("request - request.1;.csv", sep=";", header=False, index=False)
request.assign(l=-5, r=5, w=-10, c=12).to_csv("request - request.2;.csv", sep=";", header=False, index=False)
request.assign(l=-30, r=30, w=-60, c=62).to_csv("request - request.3;.csv", sep=";", header=False, index=False)
```

```python
r_tickers = set(request.iloc[:, 1].unique())
```

```python
header = ["date", "week", *search.iloc[0, 2:]]
```

```python
search1 = pd.DataFrame(search.iloc[2:].values, columns=header).drop("week", axis=1)
```

```python
search_melt = (
    pd.melt(search1, id_vars="date", var_name="company", value_name="price")
    [["company", "date", "price"]]
    .assign(company= lambda df: df.company.str.split("10").apply(operator.itemgetter(0)))
    .query("date==date")
    .drop_duplicates(["date", "company"])
)
search_melt = search_melt[search_melt.company.isin(r_tickers)]
```
