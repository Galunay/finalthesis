## фильтруем
* компании в алфавитном
* везде дата > 01.05.10

```python
ticker1 = {i: df.sort_values("company").query("date >= '2010-05-01'") for i, df in ticker.items()}
```

```python
df0 = ticker1[0].set_index(["company", "date"])
for i in range(1, 7):
    df0 = df0.join(ticker1[i].set_index(["company", "date"]), how="inner")
ticker2 = df0.reset_index().assign(quarter=lambda df: df.date.apply(lambda d: pd.datetime(d.year, (d.month-1) // 3 * 3 + 1, 1)))
```

```python
import re
def fmtB(s):
    if pd.isna(s): return
    s = s.replace(",", "")
    m = re.match(r"^\$(.+)B$", s).group(1)
    return float(m)
```

```python
ticker2m1 = ticker2.assign(quarter=lambda df: df.quarter + pd.tseries.offsets.DateOffset(months=3))
```

```python
request_ticker = (
    request1.set_index(["company", "quarter"])
    .join(ticker2.set_index(["company", "quarter"]), rsuffix="_ticker")
    .assign(NI=lambda df: df.NI.apply(fmtB), Asset=lambda df: df.Asset.apply(fmtB))
)
request_tickerm1 = (
    request1.set_index(["company", "quarter"])
    .join(ticker2m1.set_index(["company", "quarter"]), rsuffix="_ticker")
    .assign(NI=lambda df: df.NI.apply(fmtB), Asset=lambda df: df.Asset.apply(fmtB))
)
```

```python
request_ticker.columns
```

```python
diff_columns = ['total debt', 'equity', 'DERatio', 'ebitda', 'ebit',
       'eps', 'NI', 'Asset', 'ROA', 'MV', 'BV', 'PB', 'shares']
diff_out_columns = [s + "_diff" for s in diff_columns]
```

```python
request_ticker[diff_out_columns] = request_ticker[diff_columns] - request_tickerm1[diff_columns]
```

```python
request_ticker.reset_index().to_csv("request_ticker_diffs.csv")
```
